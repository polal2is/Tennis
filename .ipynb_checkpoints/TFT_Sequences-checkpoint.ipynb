```python
from pathlib import Path
import pickle
import warnings

import numpy as np
import pandas as pd
from pandas.core.common import SettingWithCopyWarning
import pytorch_lightning as pl
from pytorch_lightning.callbacks import EarlyStopping, LearningRateMonitor
from pytorch_lightning.loggers import TensorBoardLogger
import torch

from pytorch_forecasting import GroupNormalizer, TemporalFusionTransformer, TimeSeriesDataSet
from pytorch_forecasting.data.examples import get_stallion_data
from pytorch_forecasting.metrics import MAE, RMSE, SMAPE, PoissonLoss, QuantileLoss, CrossEntropy, MultiLoss
from pytorch_forecasting.models.temporal_fusion_transformer.tuning import optimize_hyperparameters
from pytorch_forecasting.utils import profile

from datetime import datetime
import pytz

Now = datetime.utcnow()
NowTZ = datetime.now(pytz.timezone("Europe/Zurich"))
TZ = NowTZ.tzinfo.utcoffset(NowTZ)
Now = Now + TZ
Now = pd.to_datetime(Now)

warnings.simplefilter("error", category=SettingWithCopyWarning)
```

```python
History = '60D'

Start_Date = (Now - pd.to_timedelta(History)).strftime('%Y-%m-%d')
print(Start_Date)

df = pd.read_csv('data.csv')

df = df.iloc[:, 1:]
df['tourney_date'] = pd.to_datetime(df['tourney_date'], infer_datetime_format = True)#, yearfirst = True, format = '%Y%m%d')

Mask_Date_Min = df['tourney_date'].dt.year >= 2010
df = df[Mask_Date_Min]

df.info()
```

```python
Min_Steps = 5

Tournament_List = df['tourney_id'].unique().tolist()

data = pd.DataFrame()
for ii, Tournament in enumerate(Tournament_List):
#     print(IEP)
    
    Mask_Tournament = df['tourney_id'] == Tournament
    
    df_tournament = df[Mask_Tournament].sort_values('tourney_date').reset_index(drop = True)
    
    len_path = len(df[Mask_Tournament])
    
    df_tournament['Rounds Played'] = len_path
    
    #if len_path >= Min_Steps:
    
    # df_tournament['Durations'] = df_tournament['tourney_date'].diff()
    # df_tournament['Durations_Hours'] = df_tournament['Durations']/pd.to_timedelta('1H')

    df_tournament['time_index'] = range(len_path)

    data = pd.concat([data, df_tournament], axis = 0, ignore_index=True)
    
# data = data.sort_values(['tourney_id', 'date_mvt']).reset_index(drop = True)
data
```

```python
Mask_Player_Wins = data['Player Wins Flag'] == 1
data['Player Age'] = np.where(Mask_Player_Wins, data['winner_age'], data['loser_age'])
data['Opponent Age'] = np.where(Mask_Player_Wins, data['loser_age'], data['winner_age'])
data['Opponent'] = np.where(Mask_Player_Wins, data['Full Name_loser'], data['Full Name_winner'])
data['Opponent Nationality'] = np.where(Mask_Player_Wins, data['Nationality_loser'], data['Nationality_winner'])
data['Player Ranking'] = np.where(Mask_Player_Wins, data['winner_rank'], data['loser_rank']).astype(int)
data['Opponent Ranking'] = np.where(Mask_Player_Wins, data['loser_rank'], data['winner_rank']).astype(int)
data['Opponent Hand'] = np.where(Mask_Player_Wins, data['loser_hand'], data['winner_hand'])
```

```python
data['tourney_id'] = data['tourney_id'].astype(str).astype("category")
data['Opponent'] = data['Opponent'].astype(str).astype("category")
data['Opponent Nationality'] = data['Opponent Nationality'].astype(str).astype("category")
data['Opponent Hand'] = data['Opponent Hand'].astype(str).astype("category")

data['tourney_year'] = data['tourney_year'].astype(str).astype("category")
data['tourney_month'] = data['tourney_month'].astype(str).astype("category")
data['surface'] = data['surface'].astype(str).astype("category")

data
```

```python
data
```

```python
max_prediction_length = 3
max_encoder_length = 4

training_cutoff = data["time_index"].max() - max_prediction_length

training = TimeSeriesDataSet(
    data[lambda x: x.time_index <= training_cutoff],
    # data,
    time_idx = "time_index",
    target = "Player Wins Flag",
    group_ids = ["tourney_id"],
    min_encoder_length = max_encoder_length // 2,  # allow encoder lengths from 0 to max_prediction_length
    # min_encoder_length = 1,  # allow encoder lengths from 0 to max_prediction_length
    max_encoder_length = max_encoder_length,
    min_prediction_length = 1,
    max_prediction_length = max_prediction_length,
    static_categoricals = ["tourney_id", "tourney_name", "tourney_level", "surface", "tourney_year", "tourney_month"],
    static_reals = ["Player Age", "Player Ranking"],
    time_varying_known_categoricals = [],
    time_varying_known_reals = ["time_idx"],
    time_varying_unknown_categoricals = [], #"Opponent Nationality", "Opponent Hand"
    time_varying_unknown_reals = ["Opponent Ranking", "Opponent Age"], #"minutes"
    # target_normalizer=GroupNormalizer(
    #     groups=["iep", "First_UF"], transformation="softplus", center=False
    # ),  # use softplus with beta=1.0 and normalize by group
    add_relative_time_idx = True,
    add_target_scales = True,
    add_encoder_length = True,
)

validation = TimeSeriesDataSet.from_dataset(training, data, predict=True, stop_randomization = True)
batch_size = 128
train_dataloader = training.to_dataloader(train=True, batch_size=batch_size, num_workers = 0)
val_dataloader = validation.to_dataloader(train=False, batch_size=batch_size, num_workers = 0)
```

```python
# save datasets
training.save("training.pkl")
validation.save("validation.pkl")
```

```python
early_stop_callback = EarlyStopping(monitor="val_loss", min_delta = 1e-4, patience = 10, verbose = True, mode = "min")
lr_logger = LearningRateMonitor()

trainer = pl.Trainer(
    max_epochs = 100,
    gpus = 1,
    weights_summary = "top",
    gradient_clip_val = 0.1,
    limit_train_batches = 30,
    # val_check_interval=20,
    # limit_val_batches=1,
    # fast_dev_run=True,
    # logger=logger,
    # profiler=True,
    callbacks = [lr_logger, early_stop_callback],
)
```

```python
tft = TemporalFusionTransformer.from_dataset(
    training,
    learning_rate=0.03,
    hidden_size=16,
    attention_head_size=1,
    dropout = 0.1,
    hidden_continuous_size = 8,
    #output_size = [1, 7],
    # loss = MultiLoss([CrossEntropy(), QuantileLoss([0.05, 0.5, 0.95])]),
    loss = CrossEntropy(),
    log_interval = 10,
    log_val_interval = 1,
    reduce_on_plateau_patience = 3,
)
print(f"Number of parameters in network: {tft.size()/1e3:.1f}k")
```

```python
# find optimal learning rate
# remove logging and artificial epoch size
tft.hparams.log_interval = -1
tft.hparams.log_val_interval = -1
trainer.limit_train_batches = 1.0
# run learning rate finder
res = trainer.tuner.lr_find(
    tft, train_dataloader=train_dataloader, val_dataloaders=val_dataloader, min_lr=1e-5, max_lr=1e2
)
print(f"suggested learning rate: {res.suggestion()}")
fig = res.plot(show=True, suggest=True)
fig.show()
tft.hparams.learning_rate = res.suggestion()
```

```python
trainer.fit(
    tft,
    train_dataloader = train_dataloader,
    val_dataloaders = val_dataloader,
)
```

```python
# make a prediction on entire validation set
preds, x, index = tft.predict(val_dataloader, return_index=True, fast_dev_run = False, return_x = True)
```
